```python
x = [0,1,2,3,4,5,6]
y = [0,2,4,6,8,10,12]

plt.figure(figsize = (4,2))    # grafiğin boyutu ayarlar 
plt.plot(x,y)
plt.show()
```

```python
x = [0,1,2,3,4,5,6]       # x = np.arange(7)
y = [0,2,4,6,8,10,12]     # y = np.arange(0,13,2)

plt.figure(figsize = (4,2))    # grafiğin boyutu ayarlar 
plt.plot(x,y,color = "orange", marker = "^")
plt.show()
```

```python
x = np.arange(7)
y = -x * 3 + 5
y2 = x * 5 -8

plt.figure(figsize = (12,6))
plt.scatter(x,y,marker = "*", color = "black", s = 150, alpha = 0.5, label = "yıldız")
plt.plot(x,y2, label = "line", marker = "^")
plt.legend()
plt.show()
```

```python
x = np.arange(100) / 10
y1 = np.sin(x)
y2 = np.cos(x)

plt.scatter(x,y1,color = "r", s = 40, alpha = 0.5, edgecolors = "b", linewidths = 10 , label = "sin(x)")
plt.plot(x,y2, label = "cos(x)")
plt.legend()
plt.show()
```

```python
np.random.seed(0)
data = {
    "Year": np.arange(2010, 2021),
    "Sells": np.random.randint(100, 500, 11),
    "Payments": np.random.randint(50, 400, 11),
    "Region": np.random.choice(["North", "South", "East", "West"], size = 11)
}
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize = (14,8))

plt.plot(df["Year"],df["Sells"],
        label="Sells",
        marker="o",
        linestyle="--",
        markerfacecolor="y",
        color="r")

plt.plot(df["Year"],df["Payments"],
        label="Payments",
        marker="*",
        linestyle="--",
        markerfacecolor="r",
        color="b")

plt.title("Annual Sales and Costs", fontsize = 18)
plt.xlabel("Year", fontsize = 14)
plt.ylabel("TL", fontsize = 14)
plt.legend()        
plt.show()
```

```python
dff = df.groupby("Region")["Sells"].sum()
dff
```

```python
plt.bar(dff.index, dff.values)
plt.title("Region")
plt.show()
```

```python
plt.bar(data = df, x = df["Region"], y = df["Sells"])
```

```python
plt.bar(dff2.index, dff2.values)
plt.title("Region")
plt.show()
```
