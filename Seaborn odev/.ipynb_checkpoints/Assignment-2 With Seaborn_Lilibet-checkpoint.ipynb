## 1a: Box plot of age in both genders separeted by manner of death.

```python
kill["manner_of_death"].unique()
```

```python
kill.manner_of_death.unique()
```

#### INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
kill.dropna(inplace=True)
#null degerleri dusurdum
```

```python
# with seaborn

#1.yol 
plt.figure(figsize=(4.3, 3))

sns.boxplot(data=kill, x="gender", y="age", hue="manner_of_death")

#icine data=kill diye vermezsem, kill[gender] seklinde cagirmam gerekirdi

plt.tick_params(axis = 'y',labelsize = 6)
#plt.yticks(fontsize=6)
plt.xlabel('gender', fontsize=7)
plt.ylabel('age', fontsize=7)

plt.tick_params(axis = 'x',labelsize = 6)
#plt.xticks(fontsize=6)

plt.legend(title = 'manner_of_death', fontsize='x-small',title_fontsize= 'x-small')

plt.show()

#soruda bizden hem kadin hem erkekler icin, manner of deathe gore ayristirilmis 
##yas datasini istiyor. quartiles of age dedigi icin boxplot
#70 yas ustu uyarisiz vurulan erkeklerde outlier fazla.
#boxplottan en cok outlier gozlemliyorum.
#kadinlarda once elektrosok verip oldurulen kadinlar genelde 25-40 arasinda.mediani 37'lerde.
#data normal dagilmamis, median yukarida kalmis, left skewed deni.
```

## 1b: Swarm Plot 

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn

plt.figure(figsize=(5, 4))

sns.swarmplot(data=kill, x="gender", y="age", hue="manner_of_death")

plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
plt.xlabel('gender', fontsize=7)
plt.ylabel('age', fontsize=7)

plt.show()

#vurularak oldurulme, uyarilarak oldurulmeden daha cok
#vurulmalarda yas yogunlugu 20-60 arasinda yogunlasmis
#vurularak oldurulmede erkekler kadinlara gore daha fazla 
#outlier'in yasi erkeklerde daha fazla
```

## 2: Plot 12 most dangerous cities

```python
city_count=kill.city.value_counts().head(12)
city_count

#df.groupby('city')['city'].count().sort_values(ascending=False).head(12) 
#bu sekilde de yapabilirsn
#df.value_counts("city").nlargest(12)
#bu sekilde de yapabilirsin.
```

```python
city_count.values
```

```python
print(kill['city'].value_counts()["Oklahoma City"])
```

```python
print(kill['city'].value_counts()["Bakersfield"])
```

```python
city_count.index
```

hem Oklohoma City hem de Bakersfield index 13, en tehlikeli sehirlerden.
ama bendeki siralamada Bakesfieledi once getiriyor.

```python
plt.figure(figsize=(8.5, 4))

sns.barplot(data=kill, x=city_count.index, y=city_count.values)  

plt.title('Most Dangerous Cities', fontsize=12, color='r')
plt.tick_params(axis='x', labelsize=7, rotation=45, colors='b')
plt.tick_params(axis='y', labelsize=7)

plt.show()

#xticks yerine tick_params kullandim,
#cunku icinde rotation parametresi var bunda.

#alternatif cozum:
#fig, ax = plt.subplots(figsize=(10,6))
#sns.barplot(x=xindex, y=yvalue)
#ax.set_xticklabels(cty.index, color="b", fontsize=8, rotation= 45)
#ax.set_title("Most Dangerous City", color="r");
```

```python
plt.figure(figsize=(6, 4))

sns.countplot(data=kill, x="city", order=city_count.index)     
#orderi yeni define ettigim city_countun indeksine gore yap dedim

plt.title('Most dangerous cities', fontsize=12, color='b')

plt.tick_params(axis='x', labelsize=7, rotation=45)
plt.tick_params(axis='y', labelsize=7)

plt.xlabel("city", fontsize=7)
plt.ylabel("count", fontsize=7)

plt.show()

#order parametresinin icine city_count index yazarsam, en cok oldurulen sehirleri siraliyor
```

## 3: Plot 20 most dangerous states

```python
state_count = kill.state.value_counts().head(20)
state_count
```

```python
plt.figure(figsize=(5.8, 4))

sns.barplot(data=kill, x=state_count.index, y=state_count.values)
#X'e eyaletlerin kisaltmasini, y'ye de degerlerini yazdirmis olduk
plt.title('Most dangerous states', fontsize=12, color='b')
plt.tick_params(axis='x', labelsize=7, rotation=45)
plt.tick_params(axis='y', labelsize=7)

plt.show()

#PA'dan sonra neden siralama karisiyor bilmiyorum?
```
