## Style

```python
plt.figure(figsize=(10,5))

sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips)
plt.show()

# hue kategorik olmalı
```

**H0** Yokluk hipotezi, Temel hipotez. "H0 reddedildi ya da edilemedi" edildiğinde H1'e geçilir.
**H1** Alternatif hipotezi

```python
fig, ax=plt.subplots(1,2, figsize=(10,5))

sns.scatterplot(x = "total_bill", y = "tip", data = tips, 
                style="sex", ax= ax[0] )
sns.scatterplot(x = "total_bill", y = "tip", data = tips, 
                hue="smoker", ax= ax[1])

plt.show()
```

## FacetGrid
### add_legend()

```python
g = sns.FacetGrid( data = tips,
                 col = "sex",
                 hue = "smoker")

g.map(sns.scatterplot, "total_bill", "tip", alpha = 0.5)
g.add_legend();
```


### shade=

```python
g = sns.FacetGrid( data = tips,
                 col = "sex",
                 hue = "smoker")

g.map(sns.kdeplot, 
      "total_bill", 
      "tip", 
      alpha = 0.5, 
      shade=True)
# fill=True da shade=True ile aynı görseli verir.

g.add_legend();
```

```python
g = sns.FacetGrid( data = tips,
                 col = "day",
                 hue = "smoker")

g.map(sns.kdeplot, 
      "total_bill", 
      "tip", 
      alpha = 0.5, 
      shade=True)

g.add_legend();
```

## Doppler Effect
### height=

```python
fig, ax = plt.subplots(1,3, figsize= (10,5))

sns.rugplot(x= "total_bill", data = tips, height=.5, ax = ax[0])
sns.histplot(x= "total_bill", data = tips, bins=50, ax = ax[1], kde=True)
sns.kdeplot(x= "total_bill", data = tips, ax = ax[2]);

# rugplot binlerin tepeden görünüşüdür.
# rugplot genelde tek başına kullanılmaz.
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))

sns.rugplot(data = tips, x = "total_bill", 
            height = 0.1, ax = ax[0])
sns.histplot(data = tips, x = "total_bill", 
             bins = 50, kde = True, ax = ax[1]);
```
